#Data Wrangling

```python
#Loading datasets - remember to drop in the survey csvs before proceeding
df = pd.read_csv("Reebok_Survey_Econded.csv")
df
```

```python
#Creating unique dataframe containing frequency - simulation
unique_df = (df.groupby(['Age', 'Gender', 'States','Region'])
                           .States
                           .agg(
                               Count = 'size'
                           )
                           .reset_index())
```
