```python
#Loading lib
import pandas as pd
import numpy as np
```

#Data Wrangling

```python
df = pd.read_csv("Twitter_Ready_Post.csv")
df
```

```python
#Generating simulation data - sentiment
df['Sentiment'] = np.random.randint(0,2, size=len(df))
df
```

```python
#Checking sentiment to make sure it is 0 and 1
print(df['Sentiment'].unique())
```

```python
#Dropping nan-rows
df = df.dropna()
df
#11204 rows total
```

```python
#dropping un-used columns
df = df.drop(['brand','org_m3'],axis = 1)
```

```python
df['gender_m3'].unique()
```

```python
df['age_m3'].unique()
```

```python
cleanup = {"gender_m3":     {"male": 1, "female": 0, "NB": 2},
                "age_m3": {"<=18": 0, "19-29": 1, "30-39": 2, ">=40": 3}}
df = df.replace(cleanup)
```

```python
#Creating unique dataframe containing frequency - simulation
unique_df = (df.groupby(['gender_m3', 'age', 'state', 'brand'])
                           .state
                           .agg(
                               n = 'size'
                           )
                           .reset_index())
```

```python
unique_df
```

```python
#Checking reduction in data size
unique_df.shape[0] / df.shape[0]

#use survey data's preference by state
```

```python
#Renaming Column
unique_df.columns = ['Gender', 'Age', 'States', 'Brand', 'Count']
```

```python
unique_df.to_csv('Twitter Post-Stratification.csv',index = False)
```
